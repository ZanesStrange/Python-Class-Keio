```python
# Your fancy code here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
cpp = pd.read_csv('CarPrice_assignment.csv')
np.cov(cpp['price'], cpp['enginesize'])
```

```python
np.corrcoef(cpp['price'], cpp['enginesize'])
```

There is a strong linear relationship between price and enginesize.

```python
sns.pairplot(cpp[['price', 'enginesize']])
plt.grid(True)
plt.show()
```

```python
#More code and comments...
sns.pairplot(cpp[['price', 'enginesize', 'wheelbase', 'horsepower', 'carheight']])
plt.grid(True)
```

```python
heatData = cpp[['price', 'enginesize', 'wheelbase', 'horsepower', 'carheight']]
sns.heatmap(heatData.corr(), annot=True)
plt.show()
```

Cars with a larger engine size and higher horsepower are likely to be more expensive based from the strong positive correlation with price. On the other hand, car height seems to have little to no correlation with other variables.

## Question 3

```python
reg = linear_model.LinearRegression()

variables = ['enginesize', 'wheelbase', 'horsepower', 'carheight']

for var in variables:
    X = cpp[[var]].values
    Y = cpp[['price']].values

    reg.fit(X, Y)

    yPred = reg.coef_ * X + reg.intercept_

    print(f"Regression for {var}")
    print(f"Slope: {reg.coef_}")
    print(f"y-Intercept: {reg.intercept_}")
    print(f"R-squared: {reg.score(X, Y)} \n")
    
    plt.scatter(X, Y)
    plt.plot(X, yPred, color='red')
    plt.grid(True)
    plt.show()
```

The regression for enginesize provides the best fit as its R-squared value is the highest among the 4 regressions.
