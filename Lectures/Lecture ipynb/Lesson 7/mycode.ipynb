```python
import random
import matplotlib.pyplot as plt

def roll_dice(n):
    faces = [1, 2, 3, 4, 5, 6]
    counts = [0] * 6

    for _ in range(n):
        r = random.randint(1, 6)
        counts[r - 1] += 1
    
    rel_freq = [c / n for c in counts]

    plt.bar(faces, rel_freq, tick_label=faces)
    plt.xlabel("Die face")
    plt.ylabel("Relative frequency")
    plt.ylim(0, 1)
    plt.title(f"Relative Frequencies for (n) Rolls")
    plt.show()

    return rel_freq

roll_dice(50000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# mean, std. dev

mu, sigma = 0, 1
xs = np.linspace(mu - 4*sigma, mu + 4*sigma, 400)
plt.figure(figsize=(6,4))
plt.plot(xs, norm.pdf(xs, mu, sigma), label=r"pdf $N(0,1)$")
plt.plot(xs, norm.cdf(xs, mu, sigma), label=r"cdf $N(0,1)$")
plt.xlabel("x")
plt.title("Normal pdf and cdf")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

def roll_dice(n):
    faces = [1, 2, 3, 4, 5, 6]
    counts = [0] * 6

    for _ in range(n):
        r = random.randint(1, 6)
        counts[r - 1] += 1
    
    rel_freq = [c / n for c in counts]

    ecdf = [sum(rel_freq)]
    plt.step(faces, ecdf, where='post', label='Empirical CDF')

    theoretical_cdf = [i / 6 for i in faces]
    plt.step(faces, theoretical_cdf, where='post', linestyle='--', label='Theoretical CDF')

    plt.xlabel("Die Face")
    plt.ylabel("CDF")
    plt.title(f"Empirical vs. Theoretical CDF ({n} rolls)")
    plt.legend()
    plt.grid(True)
    plt.show()

roll_dice(50000)
```

```python
import random
import matplotlib.pyplot as plt
from scipy.stats import norm

def roll_dice_cdf(n):
    faces = [1, 2, 3, 4, 5, 6]
    counts = [0] * 6

    for _ in range(n):
        r = random.randint(1, 6)
        counts[r - 1] += 1

    # Relative frequencies (PMF)
    rel_freq = [c / n for c in counts]

    # CDF: cumulative sum of relative frequencies
    empirical_cdf = []
    cumulative = 0
    for p in rel_freq:
        cumulative += p
        empirical_cdf.append(cumulative)

    # Plot empirical CDF
    plt.step(faces, empirical_cdf, where='post', label='Empirical CDF')

    # Theoretical CDF
    theoretical_cdf = [i / 6 for i in faces]
    plt.step(faces, theoretical_cdf, where='post', linestyle='--', label='Theoretical CDF')

    plt.xlabel("Die Face")
    plt.ylabel("CDF")
    plt.title(f"Empirical vs. Theoretical CDF ({n} rolls)")
    plt.legend()
    plt.grid(True)
    plt.show()

roll_dice_cdf(50000)
```
