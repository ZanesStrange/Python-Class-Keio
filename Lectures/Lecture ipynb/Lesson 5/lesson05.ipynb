```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
student_data = pd.read_csv('student-mat.csv', sep=';')
```

```python
# Scatter plot
plt.plot(student_data['G1'], student_data['G3'], 'o')

# Add labels
plt.xlabel('G1 grade')
plt.ylabel('G3 grade')
plt.grid(True)
plt.savefig("output_graph_7.png")
```

```python
# Covariance matrix
np.cov(student_data['G1'], student_data['G3'])
```

```python
# Correlation coeficient
np.corrcoef(student_data['G1'], student_data['G3'])
```

```python
# Coefficient of Variation
cv1 = student_data['G1'].std()/student_data['G1'].mean()
cv3 = student_data['G3'].std()/student_data['G3'].mean()
print(f"The CV for G1: {cv1*100 : .2f}%")
print(f"The CV for G3: {cv3*100 : .2f}%")
```

```python
sns_plot = sns.pairplot(student_data[['Dalc', 'Walc', 'G1', 'G3']])
plt.grid(True)
sns_plot.figure.savefig("output_graph_8.png")
```

```python
from sklearn import linear_model

# Define instance of linear regression
reg = linear_model.LinearRegression()
```

```python
# With loc we can extract all the columns and rows from a DataFrame
#Extraction of explanatory variable (must be always a [n, 2] array)
X = student_data.loc[:,['G1']].values

# Response variable
Y = student_data['G3'].values

# Call the regression method
reg.fit(X, Y)

# Coefficients for the model
print(f"Slope: {reg.coef_}")
print(f"y-Intercept: {reg.intercept_}")

```

```python
# Plot the least squares line
plt.scatter(X, Y)
plt.xlabel('G1 grade')
plt.ylabel('G3 grade')

# Over the scatter plot, we draw the LSL
plt.plot(X, reg.predict(X))
plt.grid(True)
```

```python
# Print the R^2
print(f"R-squared: {reg.score(X, Y)}")
```
